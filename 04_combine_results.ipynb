```python
from google.colab import drive
drive.mount('/content/drive')

BASE = "/content/drive/MyDrive/vision_lab"
SEQUENCES = ["nostructure_notexture", "structure_notexture", "nostructure_texture", "structure_texture"]
```

```python
import pandas as pd

all_dfs = []
for seq in SEQUENCES:
    df = pd.read_csv(f"{BASE}/{seq}/results.csv")
    df["sequence"] = seq
    all_dfs.append(df)

master = pd.concat(all_dfs, ignore_index=True)
master = master[["sequence", "method", "MAE", "AbsRel", "RMSE", "delta_1.25"]]
print(master.to_string(index=False))
```

```python
master.to_csv(f"{BASE}/master_results.csv", index=False)
print(f"Saved to {BASE}/master_results.csv")
```

```python
pivot = master[master["method"].isin(["raw", "hybrid"])].pivot(
    index="sequence", columns="method", values="MAE"
)
pivot["hybrid_improvement_%"] = (pivot["raw"] - pivot["hybrid"]) / pivot["raw"] * 100
print(pivot)
```
